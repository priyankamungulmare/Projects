# C) Logistic Regression:
**a. Build a simple logistic regression model.**

**i. Divide the dataset in 70:30 ratio.**


```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()  
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
hd[columns_to_scale] = StandardScaler.fit_transform(hd[columns_to_scale])
```

```python
hd.head()
```

```python
X = hd.drop(['target'], axis=1)
y = hd['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

**ii. Build the model on train set and predict the values on test set.**


```python
model1 = lr.fit(X_train,y_train)
prediction1 = model1.predict(X_test)
```

**iii. Build the confusion matrix and get the accuracy score.**

```python
from sklearn.metrics import confusion_matrix

cm1 = confusion_matrix(y_test, prediction1)
cm1
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, prediction1)
```

# D) Decision Tree:
**a. Build a decision tree model.**

**i. Divide the dataset in 70:30 ratio.**

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
```

**ii. Build the model on train set and predict the values on test set.**


```python
model2 = dtc.fit(X_train,y_train)
prediction2 = model2.predict(X_test)
```

**iii. Build the confusion matrix and calculate the accuracy.**


```python
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(y_test, prediction2)
cm2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction2)
```

**iv. Visualize the decision tree using the graphviz package.**

```python
feature_names = [i for i in X_train.columns]
y_train_str = y_train.astype('str')
y_train_str[y_train_str=='0']='no disease'
y_train_str[y_train_str=='1']='disease'
y_train_str=y_train_str.values
```

```python
from sklearn.tree import export_graphviz

export_graphviz(dtc, out_file="dtc.dot", feature_names=feature_names, class_names=y_train_str)
```

```python
import graphviz

with open("dtc.dot") as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```
